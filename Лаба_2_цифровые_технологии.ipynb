```python
import math as m
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
url = "https://drive.google.com/uc?export=download&id=10X7i5PYwSAe08JtXUKDBkfWUHTuI8NMm/view?usp=drive_link"
with open ('Current.csv','r') as current:  
  values = [float(line.strip()) for line in current]
  mean_value = sum(values)/len(values)
print('Среднее значение тока равно', mean_value)
resistance = 10 
with open ('Current.csv','r') as current:
  current_values = [float(line.strip()) for line in current]
  power_values = [current**2*resistance for current in current_values]
  average_power = np.mean(power_values)  
if average_power>(mean_value**2)*resistance:
  print('Среднеарифметическое значение мощности по графику больше произведения среднего тока в квадрате на сопротивление')
if average_power<(mean_value**2)*resistance:
  print('Среднеарифметическое значение мощности по графику меньше произведения среднего тока в квадрате на сопротивление ')
if average_power == (mean_value**2)*resistance:
  print('Среднеарифметическое значение мощности по графиу равно произведению среднего тока в квадрате на сопротивление')  
plt.plot(power_values)
plt.xlabel('Время')
plt.ylabel('Мощность')
plt.show()
```
